# brainana: Brain-Reward Autoresearch Agent

**An autoresearch agent where brain response IS the reward function.**

Karpathy's autoresearch optimizes `train.py` against `val_bpb`. We optimize **stimuli** against **brain activation**.

Give brainana a neural objective and it autonomously:
1. Generates text stimuli (via Claude)
2. Predicts brain response (via TRIBE v2)
3. Measures target region activation (Destrieux atlas, fsaverage5)
4. Interprets results with neuroscience reasoning
5. Iterates until convergence
6. Produces brain maps + research report

Built at the Paradigm Automated Research Hackathon, April 9, 2026.

## Run brainana with Mock Brain (instant, for testing)

```python
from brainana.agent import run_brainana

report = run_brainana(
    objective="What maximally activates the prefrontal cortex?",
    target_region="prefrontal",
    max_iterations=10,
    backend="mock",
    output_dir="./outputs/prefrontal",
)
print(f"Report: {report}")
```

## View Brain Maps

```python
from IPython.display import Image, display
from pathlib import Path

output_dir = Path('./outputs/prefrontal')

# Show score progression
display(Image(filename=str(output_dir / 'score_progression.png'), width=800))

# Show best brain map (last iteration that was best)
brain_maps = sorted(output_dir.glob('brain_iter_*.png'))
if brain_maps:
    print(f'\nBest brain map:')
    display(Image(filename=str(brain_maps[-1]), width=800))
```

## View Research Report

```python
from IPython.display import Markdown

reports = sorted(Path('./outputs/prefrontal').glob('report_*.md'))
if reports:
    display(Markdown(reports[-1].read_text()))
```
